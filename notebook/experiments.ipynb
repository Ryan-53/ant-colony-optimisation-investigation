```python
import pandas as pd
exp = pd.read_csv('experiments_ms.csv')
```

```python
exp.head()
```

**1.** Colony Size

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(figsize=(12,6))

exp.plot(ax=axs, x='colony_size',
                 y='burma_colony',
                 #grid=True,
                 label="Burma")

exp.plot(ax=axs, x='colony_size',
                 y='brazil_colony',
                 secondary_y=True,
                 label="Brazil")

plt.title('Colony size experiment')
axs.set_xlabel("Colony size")
axs.set_ylabel('Median Best Quality Found (Burma)')
plt.ylabel('Median Best Quality Found (Brazil)')
```
